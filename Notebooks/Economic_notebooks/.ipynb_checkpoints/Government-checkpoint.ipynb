## Notebook for looking at the Government.csv

```python
import pandas as pd 
```

```python
df = pd.read_csv('../../Notebooks/Datasets/Economic_dataset/Government.csv')
for col in df.columns:
    print(col)
df
```

 Drop Irrelavant Fields


```python
df2 = df.drop(columns=["Unnamed: 1","Unnamed: 2","Unnamed: 3"])
df2
```

Type of Events 

```python
Events = df2.drop(columns = ["Date Time"]).drop_duplicates("Event")
Events
```

Filter By Events

```python
event = "Prime Minister's Question Time" #Type of event
filtered_by_event = df2[df2["Event"] == event]
filtered_by_event
```

Filter By Date

```python
Day = 20
Month = 11
Year = 2001

Date = str(Month)+"/"+str(Day)+"/"+str(Year) #Type of event
filtered_by_date = df2[df2["Date Time"].str.contains(Date)]
filtered_by_date
```
